#### This notebook contains a brief explanation of the genarate_data function.

Let's begin by creating a data set. We will use the generate_data function from the utilities function file. This function takes the number data points we want to generate, and the number of clusters to group them into. 

The function returns the generated coordinates, along with their respective allocations (+1 or -1). 

```python
data, allocations = generate_data(1000, 2)
```

Let's plot the data set. The data variable simply contains a list of coordinates. data[0] will give the first pair of generated coordinates, therefore data[0, 0] gives the x coordinate, and data[0, 1] the y coordinate. 

To plot all the coordinates, we could use a for loop to iterate through each element in the list. More compactly, we can take advantage of the : (colon) argument to an array to access all the elements along a given axis, as below

```python
plt.scatter(data[:, 0], data[:, 1], s=1)
plt.show()
```

As can be seen, the function has genarated two clusters of data. If we examine the allocations array, we can see that it contains either 0 or 1. This allocation array is used to separate the data into clusters, and to check whether an algorithm has made the correct assignments or not.

Below, we've separated the data set into clusters according to the allocations array. We have then plotted these clusters in different colours.

```python
# Separate data into clusters.
cluster_0 = data[allocations == 0]
cluster_1 = data[allocations == 1]

# Plot each cluster
plt.scatter(cluster_0[:, 0], cluster_0[:, 1], c='r', s=1)
plt.scatter(cluster_0[:, 1], cluster_1[:, 1], c='b', s=1)
plt.show()
```

This is the main functionality of the generate_data function.
